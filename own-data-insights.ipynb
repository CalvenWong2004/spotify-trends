Import the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Load the CSV 

```python
df = pd.read_csv("data/blue.csv")

total_art = df["Artist Name(s)"].nunique()
total_gen = df["Genres"].nunique()

print(f"Total number of artist: {total_art}")
print(f"Total Number of Genres: {total_gen}")
```

Check for null values 

```python
missing = pd.isnull(df).sum()

print("There are some missing values in the Genres columns due to several reasons \n 1. Lack of genre detection in some songs \n 2. Genres are not specified when the songs were released\n")
print(missing)
```

```python
df.info()
```

```python
df["Genres"] = df["Genres"].astype(str).str.lower()

rules = [
    df["Genres"].str.contains("k-pop|k-ballad|k-rap", na=False),
    df["Genres"].str.contains("mandopop|c-pop|chinese rock|taiwanese pop|cantopop|chinese r&b|gufeng|malay pop", na = False),
    df["Genres"].str.contains("melodic bass|trance|progressive trance|future bass|edm|dubstep|moombahton|tropical house|chillstep|synthpop|progressive house|electronica", na = False),
    df["Genres"].str.contains("indonesian pop|indonesian r&b|indonesian indie", na=False),
    df["Genres"].str.contains("j-pop|j-rock|anime|japanese indie", na=False),
    df["Genres"].str.contains("dark r&b|r&b|trap soul|neo soul|jersey club",na=False),
    df["Genres"].str.contains("pop|dansk pop|norweigan pop|soft pop|pop singer-songwriter|hip hop|west coast hip hop|folk pop|hyperpop" , na=False),
    df["Genres"].str.contains("rap rock|emo rap|rap", na=False),
    df["Genres"].str.contains("soundtrack|country|pop country|acoustic country|spoken word|post-grunge|musicals", na=False)
]

selections = [
    "Korean", 
    "Mandarin", 
    "EDM/Electronic",
    "Indonesian",
    "Japanese",
    "R&B",
    "Pop",
    "Rap",
    "Other/Acoustic" 
]

df["Category"] = np.select(rules, selections, default = "English Pop/Unknown")

print(df["Category"].value_counts())
```

```python
# Isolate the songs that have no category
unknown_songs = df[df['Category'] == 'Unknown']

# Show the top 10 artists in that missing pile
print("--- Who is missing genre data? ---")
print(unknown_songs['Artist Name(s)'].value_counts().head(50))
```
